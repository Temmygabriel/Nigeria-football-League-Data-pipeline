# Nigeria Premier Football League (NPFL) Standings Analysis

## Objective
Scraping real NPFL league table data from the official NPFL website
to derive performance metrics and insights.

```python
# explanation:
# We use requests to fetch HTML, BeautifulSoup to parse it,
# and pandas to organize and clean the data we scrape.

import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# explanation: I keep getting 403; i.e., the website is blocking my request, so I had to use the workaround below.
# Some websites block simple bot requests, so we include a headers
# string to pretend to be a real browser visiting the page.


headers = {
    "User-Agent": (
        "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
        "AppleWebKit/537.36 (KHTML, like Gecko) "
        "Chrome/120.0 Safari/537.36"
    )
}

url = "https://npfl.com.ng/npfl-table/"
response = requests.get(url, headers=headers)

response.status_code
```

```python
# explanation:
# BeautifulSoup turns the raw HTML into a navigable structure
# so we can find the league table easily.


soup = BeautifulSoup(response.text, "html.parser")
```

```python
# explanation:
# The standings table is the first <table> on this page.


table = soup.find("table")
```

```python
# explanation:
# Headers like Pos, Club, P, W, D, L, F, A, GD, Pts


headers = [th.text.strip() for th in table.find_all("th")]
headers
```

```python
# explanation:
# Loop through the table rows to pull team performance data.


rows = table.find_all("tr")[1:]
data = []

for row in rows:
    cells = row.find_all("td")
    if len(cells) == 0:
        continue
    row_data = [cell.text.strip() for cell in cells]
    data.append(row_data)

data[:5]
```

```python
# explanation:
# Convert the scraped table into a pandas DataFrame so we can
# clean and analyze it like a spreadsheet.

df = pd.DataFrame(data, columns=headers)
df.head()
```

```python
# explanation:
# Convert the numeric columns to integers so we can calculate
# metrics like Win Percentage.


numeric_cols = ["P", "W", "D", "L", "F", "A", "GD", "Pts"]

# so we are doing that with this code 

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")

df["Win_Percentage"] = round((df["W"] / df["P"]) * 100, 2)
df.head()
```

```python
# explanation:
# I had to use my PC's full path to ensure the CSV file is saved
# where I can easily access it. On my computer, this is:
# C:\Users\USER\Documents\Python files


df.to_csv(r"C:\Users\USER\Documents\Python files\npfl_standings_2025.csv", index=False)
```

```python
# explanation:
# Sort by Win Percentage from highest to lowest and display
# the top performing teams.
df.sort_values(by="Win_Percentage", ascending=False).head(5)
```
